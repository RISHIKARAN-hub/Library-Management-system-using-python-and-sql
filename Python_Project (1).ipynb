```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Training_data_set.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['body_type'],inplace=True)
```

```python
df['Distance '].fillna(df['Distance '].median(),inplace=True)
```

```python
df['engine_power'].fillna(df['engine_power'].median(),inplace=True)
```

```python
df['seat_count'].fillna(df['seat_count'].mean(),inplace=True)
```

```python
df['door_count'].fillna(df['door_count'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Distance ']=df['Distance '].astype(int)
```

```python
df['engine_power']=df['engine_power'].astype(int)
```

```python
df['seat_count']=df['seat_count'].astype(int)
```

```python
df['door_count']=df['door_count'].astype(int)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Maker']=le.fit_transform(df['Maker'])
```

```python
df['model']=le.fit_transform(df['model'])
```

```python
df['Location']=le.fit_transform(df['Location'])
```

```python
df['Owner Type']=le.fit_transform(df['Owner Type'])
```

```python
df['transmission']=le.fit_transform(df['transmission'])
```

```python
df['fuel_type']=le.fit_transform(df['fuel_type'])
```

```python
df.dtypes
```

```python
df.boxplot(fontsize=5,figsize=(11,13))
plt.show()
```

```python
df1=df.copy()
```

```python
q1=df.quantile(0.25)
```

```python
q1
```

```python
q3=df.quantile(0.75)
```

```python
q3
```

```python
iqr=q3-q1
```

```python
iqr
```

```python
ll=q1-1.5*iqr
```

```python
ll
```

```python
ul=q3+1.5*iqr
```

```python
ul
```

```python
df_outliers_removed=df[(df>ll)&(df<ul)]
df_outliers_removed
```

```python
df_outliers_removed.boxplot(fontsize=5,figsize=(8,8))
plt.show()
```

```python
del df_outliers_removed['door_count']
```

```python
df_outliers_removed.isnull().sum()
```

```python
df_outliers_removed.dropna(subset=['Distance ','manufacture_year','Age of car','engine_displacement','engine_power','seat_count','Price'],inplace=True)
```

```python
df_outliers_removed.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
df2=df_outliers_removed.copy()
```

```python
df_stand=pd.DataFrame(ss.fit_transform(df_outliers_removed),columns=df_outliers_removed.columns)
```

```python
df_stand
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
df_stand.isnull().sum()
```

```python
df_stand.columns
```

```python
X=df_stand[['Id','Maker','model','Location','Distance ','Owner Type','manufacture_year','Age of car','engine_displacement','engine_power','Vroom Audit Rating','transmission','seat_count','fuel_type']]
y=df_stand['Price']
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
y_predicted=pd.DataFrame(model.predict(X),columns=['Y_predicted'])
```

```python
y_predicted
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_predicted)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,y_predicted)
```
